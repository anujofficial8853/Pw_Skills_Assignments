Q-1 : What is hypothesis testing in statistics?

A-1 : Hypothesis testing is a statistical method used to make inferences or decisions about a population parameter based on sample data. It involves testing an assumption (hypothesis) using a formal procedure.

Q-2 : What is the null hypothesis, and how does it differ from the alternative hypothesis?

A-2 : Null hypothesis (H₀): Assumes no effect or no difference (e.g., μ = μ₀).

Alternative hypothesis (H₁ or Ha): Opposes the null, suggesting a significant effect or difference (e.g., μ ≠ μ₀).



Q-3 : What is the significance level in hypothesis testing, and why is it important?

A-3 : The significance level (α) is the probability threshold (e.g., 0.05) for rejecting the null hypothesis. It controls the likelihood of making a Type I error (false positive).



Q-4 : What does a P-value represent in hypothesis testing?

A-4 : The P-value is the probability of obtaining test results at least as extreme as the observed data, assuming the null hypothesis is true.



Q-5 : How do you interpret the P-value in hypothesis testing?

A-5 : If P-value ≤ α, reject the null hypothesis (evidence supports H₁).

If P-value > α, fail to reject the null hypothesis (insufficient evidence).

Q-6 : What are Type I and Type II errors in hypothesis testing?



A-6 : Type I error (False Positive): Rejecting H₀ when it's actually true.

Type II error (False Negative): Failing to reject H₀ when it's actually false.

Q-7 : What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

A-7 : One-tailed test: Tests for effect in one direction (e.g., μ > μ₀ or μ < μ₀).

Two-tailed test: Tests for effect in either direction (e.g., μ ≠ μ₀).



Q-8 : What is the Z-test, and when is it used in hypothesis testing?

A-8 : The Z-test is used when the population standard deviation is known and the sample size is large (n ≥ 30), assuming the data is normally distributed.

Q-10 : What is the T-distribution, and when should it be used instead of the normal distribution?

A-10 : The T-distribution is used when the sample size is small (n < 30) and the population standard deviation is unknown. It’s wider and has heavier tails than the normal distribution.



Q-11 :  What is the difference between a Z-test and a T-test?

A-11 : Z-test: Known population standard deviation, large sample.

T-test: Unknown population standard deviation, small sample.



Q-12 :  What is the T-test, and how is it used in hypothesis testing?

A-12 : The T-test compares means between groups or against a known value, particularly when the sample size is small and population standard deviation is unknown.

Q-13 : What is the relationship between Z-test and T-test in hypothesis testing?

A-13 : Both test hypotheses about means. The Z-test is a special case of the T-test where the population standard deviation is known or the sample size is large.



Q-14 : What is a confidence interval, and how is it used to interpret statistical results?

A-14 : A confidence interval (CI) is a range of values that likely contains the population parameter. For example, a 95% CI means there's 95% confidence the true value lies in that interval.



Q-15 : What is the margin of error, and how does it affect the confidence interval?

A-15 : The margin of error is the range added/subtracted to the point estimate to create the confidence interval. A larger margin means a wider interval and less precision.



Q-16 : How is Bayes' Theorem used in statistics, and what is its significance?

A-16 : Bayes’ Theorem updates the probability of a hypothesis based on new evidence.
P(A|B) = [P(B|A) × P(A)] / P(B)
It’s crucial in fields like machine learning, diagnostics, and decision-making under uncertainty.



Q-17 : What is the Chi-square distribution, and when is it used?

A-17 : The Chi-square (χ²) distribution is used for categorical data, especially in tests of independence and goodness of fit. It is skewed right and only takes positive values.

Q-18 : What is the Chi-square goodness of fit test, and how is it applied?

A-18 : It tests whether an observed frequency distribution differs from an expected one.
χ² = Σ[(O - E)² / E], where O = observed, E = expected.

Q-19 : What is the F-distribution, and when is it used in hypothesis testing?

A-19 : The F-distribution is used to compare variances. It’s the basis for the ANOVA test and F-tests. It's always positive and skewed right.



Q-20 : What is an ANOVA test, and what are its assumptions?

A-20 : ANOVA (Analysis of Variance) tests whether there are significant differences between the means of three or more groups.
Assumptions:

Independence of observations

Normal distribution within groups

Equal variances (homogeneity of variance)



Q-21 : What are the different types of ANOVA tests?

A-21 : One-way ANOVA: One independent variable

Two-way ANOVA: Two independent variables

Repeated measures ANOVA: Same subjects across multiple conditions

Q-22 : What is the F-test, and how does it relate to hypothesis testing?

A-22 : An F-test compares the variances of two or more groups. It’s used in ANOVA and regression analysis to determine if a group of variables significantly predicts the outcome.

Q-14 : Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std):
    z = (x - mean) / std
    return z

# Example data
data_point = 75
population_mean = 70
population_std = 5

# Calculate Z-score
z_score = calculate_z_score(data_point, population_mean, population_std)
print(f"Z-score for x = {data_point} is: {z_score:.2f}")

# Plot standard normal distribution
x_vals = np.linspace(-4, 4, 1000)
pdf_vals = norm.pdf(x_vals)

plt.plot(x_vals, pdf_vals, label='Standard Normal Distribution', color='blue')
plt.fill_between(x_vals, 0, pdf_vals, where=(x_vals <= z_score), color='orange', alpha=0.5, label=f"P(Z ≤ {z_score:.2f})")

# Mark the Z-score on the curve
plt.axvline(z_score, color='red', linestyle='--', label=f'Z = {z_score:.2f}')
plt.title("Z-score Comparison with Standard Normal Distribution")
plt.xlabel("Z")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

Q-15 : Implement hypothesis testing using Z-statistics for a sample dataset.

```python
import numpy as np
from scipy.stats import norm

# Sample data (e.g., heights in cm)
sample_data = np.array([168, 172, 170, 165, 169, 171, 173, 167, 174, 170])

# Known population parameters
population_mean = 170
population_std = 4  # Population standard deviation is known

# Sample statistics
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

# Z-statistic formula
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Two-tailed test: Calculate p-value
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Significance level
alpha = 0.05

# Output
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-score: {z_score:.3f}")
print(f"P-value: {p_value:.4f}")

# Hypothesis Test Result
if p_value < alpha:
    print("Reject the null hypothesis (H₀). There is a significant difference.")
else:
    print("Fail to reject the null hypothesis (H₀). No significant difference.")
```

Q-16 : Create a confidence interval for a dataset using Python and interpret the result.

```python
import numpy as np
from scipy.stats import norm

# Example dataset (e.g., student test scores)
data = np.array([78, 82, 85, 88, 75, 80, 79, 84, 81, 77])

# Known population standard deviation (if unknown, use t-distribution instead)
population_std = 5

# Confidence level
confidence = 0.95
alpha = 1 - confidence

# Sample statistics
sample_mean = np.mean(data)
sample_size = len(data)
z_critical = norm.ppf(1 - alpha/2)  # Z value for 95% confidence ≈ 1.96

# Margin of Error
margin_of_error = z_critical * (population_std / np.sqrt(sample_size))

# Confidence Interval
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Output
print(f"Sample Mean = {sample_mean:.2f}")
print(f"Z-critical value = {z_critical:.3f}")
print(f"Margin of Error = {margin_of_error:.2f}")
print(f"95% Confidence Interval = ({ci_lower:.2f}, {ci_upper:.2f})")
```

Q-17 : Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Step 1: Generate data from a normal distribution
np.random.seed(42)  # For reproducibility
mu = 100            # Population mean
sigma = 15          # Population std deviation
n = 50              # Sample size

data = np.random.normal(loc=mu, scale=sigma, size=n)

# Step 2: Calculate confidence interval
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Use sample std (ddof=1)
confidence = 0.95
z_critical = norm.ppf(1 - (1 - confidence)/2)

margin_of_error = z_critical * (sample_std / np.sqrt(n))
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Step 3: Display results
print(f"Sample Mean = {sample_mean:.2f}")
print(f"Sample Standard Deviation = {sample_std:.2f}")
print(f"Z-critical (95% CI) = {z_critical:.2f}")
print(f"Margin of Error = {margin_of_error:.2f}")
print(f"95% Confidence Interval for the Mean = ({ci_lower:.2f}, {ci_upper:.2f})")

# Step 4: Visualization
plt.hist(data, bins=10, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='--', label='Sample Mean')
plt.axvline(ci_lower, color='green', linestyle='--', label='95% CI Lower Bound')
plt.axvline(ci_upper, color='green', linestyle='--', label='95% CI Upper Bound')
plt.title("Histogram of Normally Distributed Sample Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.show()
```

Q-24 : Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis

def analyze_normal_distribution_skew_kurtosis(mean=0, std=1, size=1000, visualize=True):
    # Generate data from a normal distribution
    data = np.random.normal(loc=mean, scale=std, size=size)

    # Calculate skewness and kurtosis
    skewness = skew(data)
    kurt = kurtosis(data, fisher=False)        # Set fisher=False to get kurtosis (not excess kurtosis)
    excess_kurt = kurtosis(data, fisher=True)  # Excess kurtosis

    # Output
    print(f"Skewness: {skewness:.4f}")
    print(f"Kurtosis: {kurt:.4f} (Excess Kurtosis: {excess_kurt:.4f})")

    # Visualization (optional)
    if visualize:
        plt.hist(data, bins=30, color='lightgreen', edgecolor='black', density=True)
        plt.title("Histogram of Normally Distributed Data")
        plt.xlabel("Value")
        plt.ylabel("Density")
        plt.grid(True)
        plt.show()

# Example usage
analyze_normal_distribution_skew_kurtosis()
```

Q-1 : Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results.

```python
import numpy as np
from scipy.stats import norm

# --- Step 1: Sample and population data ---
population_mean = 100      # Known population mean (μ)
population_std = 15        # Known population standard deviation (σ)
sample = np.array([105, 98, 102, 97, 99, 101, 103, 100, 104, 98,
                   106, 100, 102, 101, 97, 99, 98, 103, 100, 102,
                   101, 100, 97, 105, 98, 100, 103, 104, 99, 101,
                   98, 96, 97, 100, 102, 101, 100, 98, 97, 99,
                   100, 103, 99, 102, 98, 101, 100, 104, 97, 96])  # Sample of 50

sample_mean = np.mean(sample)
n = len(sample)

# --- Step 2: Compute Z-score ---
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))

# --- Step 3: Compute P-value (two-tailed test) ---
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# --- Step 4: Interpretation ---
alpha = 0.05  # Significance level

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference between sample and population mean.")
```

Q-3 :  Implement a one-sample Z-test using Python to compare the sample mean with the population mean.

```python
import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample, population_mean, population_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)

    # Step 1: Calculate Z-score
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    # Step 2: Calculate P-value (two-tailed)
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    # Step 3: Interpret the result
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-score: {z_score:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print("🔍 Result: Reject the null hypothesis (significant difference).")
    else:
        print("✅ Result: Fail to reject the null hypothesis (no significant difference).")

# Example usage
np.random.seed(42)
sample_data = np.random.normal(loc=102, scale=15, size=50)  # Sample with true mean ~102
known_population_mean = 100
known_population_std = 15

one_sample_z_test(sample_data, known_population_mean, known_population_std)
```

Q-4 : Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# --- Step 1: Parameters ---
population_mean = 100     # μ
population_std = 15        # σ (known)
sample_size = 50
alpha = 0.05               # Significance level

# --- Step 2: Generate sample data ---
np.random.seed(0)
sample = np.random.normal(loc=104, scale=population_std, size=sample_size)
sample_mean = np.mean(sample)

# --- Step 3: Calculate Z-score and P-value ---
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
p_value = 2 * (1 - norm.cdf(abs(z_score)))  # Two-tailed test

# --- Step 4: Output ---
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")
if p_value < alpha:
    print("🔍 Reject the null hypothesis: significant difference.")
else:
    print("✅ Fail to reject the null hypothesis: no significant difference.")

# --- Step 5: Visualization ---
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

# Critical Z-values for 2-tailed test
z_crit = norm.ppf(1 - alpha / 2)

plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Standard Normal Distribution', color='blue')

# Shade rejection regions
plt.fill_between(x, y, where=(x <= -z_crit), color='red', alpha=0.5, label='Rejection Region (left)')
plt.fill_between(x, y, where=(x >= z_crit), color='red', alpha=0.5, label='Rejection Region (right)')

# Draw vertical line for calculated z-score
plt.axvline(z_score, color='black', linestyle='--', label=f'Z-score = {z_score:.2f}')
plt.axvline(-z_crit, color='gray', linestyle=':', label=f'Z-critical = ±{z_crit:.2f}')
plt.axvline(z_crit, color='gray', linestyle=':')

# Labels and legend
plt.title("Two-Tailed Z-Test: Decision Regions")
plt.xlabel("Z-value")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

Q-5 :  Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_type1_type2_errors(mu0=100, mu1=105, sigma=10, n=30, alpha=0.05):
    """
    Visualize Type I and Type II errors in hypothesis testing.

    Parameters:
    - mu0: Population mean under H0 (null hypothesis)
    - mu1: True mean under H1 (actual population mean)
    - sigma: Population standard deviation
    - n: Sample size
    - alpha: Significance level (Type I error rate)
    """

    # Standard error
    se = sigma / np.sqrt(n)

    # Critical value for two-tailed test
    z_crit = norm.ppf(1 - alpha/2)

    # Critical sample means under H0
    lower_crit = mu0 - z_crit * se
    upper_crit = mu0 + z_crit * se

    # Create x range to plot distributions
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)

    # Distributions
    h0_pdf = norm.pdf(x, mu0, se)
    h1_pdf = norm.pdf(x, mu1, se)

    plt.figure(figsize=(12, 6))

    # Plot null and alternative distributions
    plt.plot(x, h0_pdf, label="H₀ Distribution (μ₀)", color="blue")
    plt.plot(x, h1_pdf, label="H₁ Distribution (μ₁)", color="green")

    # Shade Type I error (alpha): reject H₀ when it's true
    plt.fill_between(x, h0_pdf, 0, where=(x < lower_crit) | (x > upper_crit),
                     color='red', alpha=0.4, label="Type I Error (α)")

    # Shade Type II error (beta): fail to reject H₀ when H₁ is true
    plt.fill_between(x, h1_pdf, 0, where=(x >= lower_crit) & (x <= upper_crit),
                     color='orange', alpha=0.4, label="Type II Error (β)")

    # Lines for critical values
    plt.axvline(lower_crit, color="black", linestyle="--", label=f"Lower Critical Value: {lower_crit:.2f}")
    plt.axvline(upper_crit, color="black", linestyle="--", label=f"Upper Critical Value: {upper_crit:.2f}")

    plt.title("Visualization of Type I and Type II Errors")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
plot_type1_type2_errors(mu0=100, mu1=105, sigma=10, n=30, alpha=0.05)
```

Q-6 : Write a Python program to perform an independent T-test and interpret the results.

```python
import numpy as np
from scipy.stats import ttest_ind

# --- Step 1: Generate sample data for two independent groups ---
np.random.seed(42)
group_a = np.random.normal(loc=100, scale=10, size=30)  # Mean = 100
group_b = np.random.normal(loc=105, scale=12, size=35)  # Mean = 105

# --- Step 2: Perform independent T-test ---
t_stat, p_value = ttest_ind(group_a, group_b, equal_var=False)  # Welch’s T-test (unequal variances)

# --- Step 3: Interpret the results ---
alpha = 0.05

print("Independent Two-Sample T-test")
print("-" * 35)
print(f"Group A Mean: {np.mean(group_a):.2f}")
print(f"Group B Mean: {np.mean(group_b):.2f}")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("🔍 Result: Reject the null hypothesis — the means are significantly different.")
else:
    print("✅ Result: Fail to reject the null hypothesis — no significant difference.")
```

Q-9 : Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    """
    Calculates the confidence interval for the sample mean.

    Parameters:
        data (list or np.array): Sample data.
        confidence (float): Confidence level (default 0.95 for 95%).

    Returns:
        (lower_bound, upper_bound): Tuple of confidence interval bounds.
    """
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    sem = np.std(data, ddof=1) / np.sqrt(n)  # Standard Error of the Mean (SEM)
    t_crit = t.ppf((1 + confidence) / 2, df=n - 1)  # t critical value
    margin_of_error = t_crit * sem
    return (mean - margin_of_error, mean + margin_of_error)

# --- Example usage ---
np.random.seed(0)
sample = np.random.normal(loc=100, scale=15, size=30)  # Simulated sample
ci = confidence_interval(sample, confidence=0.95)
print(f"95% Confidence Interval for the Mean: {ci}")
```

Q-10 : Write a Python program to calculate the margin of error for a given confidence level using sample data.

```python
import numpy as np
from scipy.stats import t

def margin_of_error(data, confidence=0.95):
    """
    Calculate the margin of error for a given sample and confidence level.

    Parameters:
        data (list or np.array): The sample data.
        confidence (float): Confidence level (default is 0.95).

    Returns:
        float: The margin of error.
    """
    data = np.array(data)
    n = len(data)
    sample_std = np.std(data, ddof=1)  # sample standard deviation
    sem = sample_std / np.sqrt(n)      # standard error of the mean
    t_critical = t.ppf((1 + confidence) / 2, df=n - 1)
    margin_error = t_critical * sem
    return margin_error

# --- Example usage ---
np.random.seed(1)
sample_data = np.random.normal(loc=70, scale=8, size=25)

moe = margin_of_error(sample_data, confidence=0.95)
print(f"Margin of Error at 95% confidence: {moe:.3f}")
```

Q-11 : E Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

```python
def bayes_theorem(prior, likelihood, false_positive_rate):
    """
    Implements Bayes' Theorem for binary classification (e.g., medical test)

    Parameters:
    - prior: P(Disease), the prior probability of the hypothesis
    - likelihood: P(Positive | Disease), the true positive rate
    - false_positive_rate: P(Positive | No Disease), the false positive rate

    Returns:
    - posterior: P(Disease | Positive)
    """
    # Total probability of a positive test (marginal likelihood)
    p_positive = (likelihood * prior) + (false_positive_rate * (1 - prior))

    # Bayes' Theorem
    posterior = (likelihood * prior) / p_positive
    return posterior

# Parameters
P_disease = 0.01
P_positive_given_disease = 0.95
P_positive_given_no_disease = 0.05

# Compute posterior probability
posterior = bayes_theorem(P_disease, P_positive_given_disease, P_positive_given_no_disease)
print(f"Probability of having the disease given a positive test: {posterior:.4f}")
```

Q-12 : Perform a Chi-square test for independence between two categorical variables in Python.

```python
import numpy as np
from scipy.stats import chi2_contingency

# Contingency table (rows: gender, columns: preference)
#           Likes  Dislikes
# Male      [20,   30]
# Female    [25,   25]
data = np.array([[20, 30],
                 [25, 25]])

# Perform Chi-square test
chi2, p, dof, expected = chi2_contingency(data)

# Display results
print("Chi-square Statistic:", round(chi2, 4))
print("Degrees of Freedom:", dof)
print("P-value:", round(p, 4))
print("Expected Frequencies Table:\n", expected)

# Interpret the result
alpha = 0.05
if p < alpha:
    print("\n❌ Reject the null hypothesis: Variables are dependent (associated).")
else:
    print("\n✅ Fail to reject the null hypothesis: Variables are independent.")
```

Q-13 : Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
data.

```python
import numpy as np
import pandas as pd

def calculate_expected_frequencies(observed):
    """
    Calculate expected frequencies for a chi-square test of independence.

    Parameters:
        observed (2D list or array): Observed frequency table (contingency table)

    Returns:
        expected (2D numpy array): Expected frequency table
    """
    observed = np.array(observed)
    row_totals = observed.sum(axis=1, keepdims=True)
    col_totals = observed.sum(axis=0, keepdims=True)
    grand_total = observed.sum()

    expected = (row_totals @ col_totals) / grand_total
    return expected

# --- Example ---
# Contingency table (rows: gender, columns: preference)
#       Likes  Dislikes
# Male     20        30
# Female   25        25
observed_data = [[20, 30],
                 [25, 25]]

expected_freq = calculate_expected_frequencies(observed_data)

# Display results in a DataFrame for clarity
df_expected = pd.DataFrame(expected_freq, columns=["Likes", "Dislikes"], index=["Male", "Female"])
print("Expected Frequencies:")
print(df_expected.round(2))
```

Q-14 :  Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

```python
import numpy as np
from scipy.stats import chisquare

# Observed frequencies from rolling a die 60 times
observed = np.array([8, 9, 10, 12, 11, 10])

# Expected frequencies for a fair die (uniform distribution)
expected = np.array([10, 10, 10, 10, 10, 10])

# Perform Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Output results
print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("\n❌ Reject the null hypothesis: The observed distribution is significantly different from the expected.")
else:
    print("\n✅ Fail to reject the null hypothesis: The observed distribution fits the expected distribution.")
```
