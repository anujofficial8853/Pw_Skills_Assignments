5. What are iterators in Python and how do they differ from iterables?

Answer :

Iterable: An object that can be looped over (like a list, tuple, string).
→ It has a built-in method called __iter__().

Iterator: An object that remembers where it is during iteration.
→ It has two methods: __iter__() and __next__(). → You can get the next item one at a time using next().

In short:

Iterable = something you can get an iterator from.

Iterator = something that gives you one item at a time.

5.  Explain the concept of generators in Python and how they are defined.

Answer :

A generator is a special type of iterator.

It generates values one by one as you loop through it — it doesn’t store everything in memory (makes it very memory-efficient!).

Instead of return, it uses the yield keyword to produce a value.

How to define a generator?

You define a generator like a normal function, but use yield instead of return.

Or you can create it quickly using generator expressions (like a shortcut for small generators).

```python
# Defining a generator function
def count_up_to(max):
    count = 1
    while count <= max:
        yield count
        count += 1

# Using the generator
counter = count_up_to(3)

print(next(counter))  # Output: 1
print(next(counter))  # Output: 2
print(next(counter))  # Output: 3
```

6. What are the advantages of using generators over regular functions?

Answer : Advantages of Generators:

Memory Efficiency:

Generators are lazy. They produce items one at a time, so they don't store everything in memory.

Useful when working with large datasets (like reading a large file or processing a big collection) without running into memory issues.

Performance:

Since they generate values only when needed, they can be more efficient in time and space, especially for large loops or large datasets.

Example : Without Generator (using a regular function)

```python
def get_even_numbers(max):
    even_numbers = []
    for num in range(1, max+1):
        if num % 2 == 0:
            even_numbers.append(num)
    return even_numbers

# Calling the function
even_nums = get_even_numbers(1000000)
print(even_nums[:5])  # Just printing the first 5 even numbers
```

With Generator:

```python
def get_even_numbers(max):
    for num in range(1, max+1):
        if num % 2 == 0:
            yield num

# Calling the generator
even_nums = get_even_numbers(1000000)
print(next(even_nums))  # Output: 2
print(next(even_nums))  # Output: 4
```

9.  Explain the purpose and usage of the `map()` function in Python.

The map() function is used to apply a function to every item in an iterable (like a list or tuple) and return a new iterable (usually a map object) with the results.

It allows you to avoid using a loop for transformations or computations on a collection.

Syntax: map(function, iterable)

function: The function you want to apply.

iterable: The iterable (e.g., list, tuple) to which the function will be applied.


 When to use map()?

Transformation tasks: When you need to apply the same operation to each element in a collection.

It's typically used in functional programming for transformations like converting data types, performing calculations, etc.



Example of map():

```python
# List of numbers
numbers = [1, 2, 3, 4, 5]

# Function to square a number
def square(x):
    return x ** 2

# Using map to apply square function to each number
squared_numbers = map(square, numbers)

# Convert the map object to a list and print
print(list(squared_numbers))
```

 Lambda with map() (shortcut):

```python
# Using lambda inside map
squared_numbers = map(lambda x: x ** 2, numbers)

# Convert the map object to a list and print
print(list(squared_numbers))
```

10.  What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

map() function:

Purpose:

Applies a function to every item in an iterable (like a list or tuple) and returns an iterable (usually a map object) of the results.

Output: An iterable with the transformed values.

Example :

```python
numbers = [1, 2, 3, 4]
result = map(lambda x: x * 2, numbers)
print(list(result))  # Output: [2, 4, 6, 8]
```

filter() function:

Purpose:

Applies a function to every item in an iterable and filters out items for which the function returns False. Only elements that satisfy the condition (i.e., return True) are kept.

Output: An iterable containing only the items that satisfy the condition.

Example :

```python
numbers = [1, 2, 3, 4, 5]
result = filter(lambda x: x % 2 == 0, numbers)
print(list(result))  # Output: [2, 4]
```

reduce() function:

Purpose:

Reduces a sequence of items to a single value by applying a function that cumulatively combines the items. This function is from the functools module.

Output: A single value after applying the function to all elements.

Example

```python
from functools import reduce

numbers = [1, 2, 3, 4]
result = reduce(lambda x, y: x + y, numbers)
print(result)  # Output: 10 (1 + 2 + 3 + 4)
```

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
def sum_of_even_numbers(numbers):
    # Use filter() to get only even numbers and then sum them up
    even_numbers = filter(lambda x: x % 2 == 0, numbers)
    return sum(even_numbers)

# Example usage:
numbers = [47, 11, 42, 13, 8, 20]
result = sum_of_even_numbers(numbers)
print(f"Sum of even numbers: {result}")
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    # Check for numbers less than 2, which are not prime
    if n <= 1:
        return False

    # Check for factors from 2 to the square root of n
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False  # Not a prime if divisible by any number other than 1 and itself

    return True  # It is prime if no divisors were found

# Example usage:
for number in range(1, 201):
    if is_prime(number):
        print(number, end=" ")
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms  # Number of terms in the Fibonacci sequence
        self.a, self.b = 0, 1  # Initializing first two terms (Fibonacci: 0, 1)
        self.count = 0  # Counter to track number of terms generated

    def __iter__(self):
        return self  # Return the iterator object itself

    def __next__(self):
        if self.count < self.terms:
            # Generate next Fibonacci number
            fib_number = self.a
            self.a, self.b = self.b, self.a + self.b  # Update a and b for the next Fibonacci number
            self.count += 1  # Increment the count
            return fib_number
        else:
            raise StopIteration  # Stop iteration after the specified number of terms

# Example usage:
fibonacci = FibonacciIterator(10)  # Create an iterator for 10 terms of Fibonacci sequence

# Using the iterator to get Fibonacci numbers
for number in fibonacci:
    print(number, end=" ")
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_2(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

# Example usage:
for power in powers_of_2(5):  # Powers of 2 up to 2^5
    print(power, end=" ")
```

7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_line_by_line(file_path):
    try:
        with open(file_path, 'r') as file:
            # Yield each line in the file one by one
            for line in file:
                yield line.strip()  # Using strip() to remove leading/trailing whitespaces
    except FileNotFoundError:
        print(f"The file at {file_path} was not found.")
    except IOError:
        print(f"An error occurred while reading the file at {file_path}.")

# Example usage:
file_path = 'example.txt'  # Replace with your file path
for line in read_file_line_by_line(file_path):
    print(line)
```

9. 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.



```python
# List of temperatures in Celsius
celsius = [0, 10, 20, 30, 40]

# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32

# Using map() to apply the conversion
fahrenheit = list(map(celsius_to_fahrenheit, celsius))

print("Temperatures in Fahrenheit:", fahrenheit)
```
